```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('adult.data.csv')
df.head()
```

# Tampilkan Plot Education berdasarkan Salarynya

```python
df['salary'].unique()
```

```python
df['education'].unique()
```

```python
education_salary = df.groupby('education')['salary'].value_counts().unstack().fillna(0)
education_salary.plot(kind='bar', stacked=True)
plt.title('Education vs Salary')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.show()
```

### Tampilkan Jumlah Pendidikan Bachelors dengan Gaji <=50k

```python
bachelors_count = df[(df['education'] == 'Bachelors') & (df['salary'] == '<=50K')].shape[0]
print(bachelors_count)
```

```python
occupation_above_50k = df[df['salary'] == '>50K']['occupation'].unique()
print(occupation_above_50k)
```

###

```python
occupation_salary = df[df['salary'] == '>50K']['occupation'].value_counts()
occupation_salary.plot(kind='bar')
plt.title('Occupation vs Salary > 50K')
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.show()
```
