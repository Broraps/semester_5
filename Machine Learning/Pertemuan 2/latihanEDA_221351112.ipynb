```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("adult.data.csv")
df.head()
```

```python
df.info()
```

```python
numerical_stats = df.describe()
print("Statistik Deskriptif Untuk Kolom Numerik :\n",numerical_stats)
```

## 3. Mengenal EDA dengan Data Numerik

```python
plt.figure(figsize=(8,6))
sns.histplot(df['age'],kde=True)
plt.title("Histogram of Age")
plt.xlabel('age')
plt.ylabel('frequency')
plt.show
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data = df, x = 'sex', y = 'hours-per-week')
plt.title('Box Plot of Hours per Week by Sex')
plt.xlabel('Sex')
plt.ylabel('Hours per week')
plt.show
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data = df, x = 'age', y = 'capital-gain')
plt.title('Scatter plot of age vs capital gain')
plt.xlabel('age')
plt.ylabel('capital gain')
plt.show()
```

## 4. Mengenal EDA dengan Data Kategori

```python
avg_capital_gain_by_age = df.groupby('age')['capital-gain'].mean()
plt.figure(figsize=(8,6))
plt.plot(avg_capital_gain_by_age.index, avg_capital_gain_by_age.values)
plt.title('Line Plot of Average Capital Gain By age')
plt.xlabel('Age')
plt.ylabel('Average Capital Gain')
plt.show()
```

```python
education_counts = df['education'].value_counts()
plt.figure(figsize=(8,6))
plt.bar(education_counts.index, education_counts.values)
plt.title('Jumlah Orang Berdasarkan Tingkat Pendidikan')
plt.xlabel('Tingkat Pendidikan')
plt.ylabel('Jumlah Orang')
plt.xticks(rotation = 45, ha = 'right')
plt.show()
```

```python
sex_counts = df['sex'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(sex_counts.values, labels=sex_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Persentase Jenis Kelamin')
plt.show()
```

```python
race_counts = df['race'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(race_counts.values, labels=race_counts, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.4))
plt.title('Persentase Ras')
plt.show()
```

## 5. Mengenal Bentuk EDA dengan Data Multivariabel

```python
numerical_features = ['age', 'fnlwgt', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week']
heatmap_data = df[numerical_features].corr()

plt.figure(figsize=(10,8))
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(df[['age','education-num', 'hours-per-week', 'capital-gain', 'capital-loss']])
plt.show()
```

```python
bubble_data = df[['age', 'hours-per-week', 'capital-gain']]

plt.figure(figsize=(10,8))
plt.scatter(data=bubble_data, x = 'age', y = 'hours-per-week', s = 'capital-gain', alpha = 0.5)
plt.xlabel('Age')
plt.ylabel('Hours per week')
plt.title('Bubble Chart : Age, Hours per week and Capital Gain')
plt.show()
```

## 6. Mengenal Univarate Analysis

```python
plt.figure(figsize=(8,6))
sns.histplot(df['age'], kde=True)
plt.title('Distribution of age')
plt.xlabel('age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x = 'education')
plt.title('Count of Individuals by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.xticks(rotation=45, ha = 'right')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x = 'salary')
plt.title('Count of Individuals by Salary')
plt.xlabel('Salary')
plt.ylabel('Count')
plt.xticks(rotation=45, ha = 'right')
plt.show()
```

## 7. Mengenal Bivariate Analysis

```python
plt.figure(figsize=(8,6))
sns.boxplot(data = df, x ='salary', y = 'age')
plt.title('Hubungan Antara umum dan tingkat pendapatan')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data = df, x ='education', hue = 'salary')
plt.title('Hubungan Antara Tingkat pendidikan dan tingkat pendapatan')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(data = df, x ='salary', hue = 'hours-per-week')
plt.title('Hubungan Antara jam kerja per minggu dan tingkat pendapatan')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(data = df, x ='sex', hue = 'salary')
plt.title('Perbedaan tingkat pendapatan berdasarkan jenis kelamin')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data = df, x ='occupation', hue = 'salary')
plt.title('Hubungan antara jenis pekerjaan dan tingkat pendapatan')
plt.xticks(rotation = 45, ha = 'right')
plt.show()
```

```python
numeric_cols = df.select_dtypes(include = ['number']).columns
correlation_matrix = df[numeric_cols].corr()
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot = True, cmap = 'coolwarm', fmt = '.2f')
plt.title('Correlation Matrix')
plt.show()
```
